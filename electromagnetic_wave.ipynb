```python
from numpy import sin, linspace, pi
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# --- Parameters --- 
E0 = 1.0       # Electric field amplitude
# B0 = 1.0       # Magnetic field amplitude
λ = 2.0        # Wavelength
k = 2 * pi / λ  # Wave number
frames = 16
ω = 2 * pi / frames  # Angular frequency (set period = 2 s)
z = linspace(0, 2 * λ, 100)  # Spatial domain

# --- Figure setup ---
fig, ax = plt.subplots(figsize=(8, 4))
ax.set_xlim(0, 2 * λ)
ax.set_ylim(-1.2, 1.2)
ax.set_xlabel("z (propagation direction)")
ax.set_ylabel("Field amplitude")
ax.set_title("Electromagnetic Wave Propagation")

# --- Lines for E and B fields ---
E_line, = ax.plot([], [], 'r-', lw=2, label='Electric Field (E)')
# B_line, = ax.plot([], [], 'b-', lw=2, label='Magnetic Field (B)')
ax.legend(loc='upper right')

# --- Initialization ---
def init():
    E_line.set_data([], [])
    # B_line.set_data([], [])
    return E_line, # B_line

# --- Animation function ---
def animate(t):
    E = E0 * sin(k * z - ω * t)
    # B = B0 * sin(k * z - ω * t)
    E_line.set_data(z, E)
    # B_line.set_data(z, B)
    return E_line, # B_line

# --- Create animation ---
ani = animation.FuncAnimation(fig, animate, frames=frames, init_func=init,
                              interval=50, blit=True)

# --- Save as video ---
ani.save("electromagnetic_wave.gif", writer='pillow', fps=20, dpi=150)

plt.show()
```
